#### NULL values in jobcity and other columns are imputed by -1. Now repalce then with 0 and others

```python
lis = ["ComputerProgramming","ElectronicsAndSemicon",'ComputerScience','MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg']
```

```python
df
```

```python
df["JobCity"] = df["JobCity"].replace("-1","others") #replacing -1 in jobcity column with others
```

### type-casting Datatypes

```python
df["DOJ"] = pd.to_datetime(df["DOJ"], format="%m/%d/%y")
```

```python
df["DOJ"].dtype
```

```python
df["DOJ"]
```

```python
df["DOB"] = pd.to_datetime(df["DOB"])
```

```python
df["DOB"] = pd.to_datetime(df["DOB"], format="%m/%d/%y")
```

### considering present as todays date in DOL columns

```python
df["DOL"] = df["DOL"].replace("present", "2/22/24 0:00")    #2/19/90 0:00
```

```python
df["DOL"] = pd.to_datetime(df["DOL"])
```

```python
df["DOL"] = pd.to_datetime(df["DOL"],format="%m/%d/%y")
```

```python
df.info()
```

### renaming columns

```python
df.columns = [col.strip().lower()   for col in df.columns]
```

```python
df.columns
```

### creating new columns

```python
df["doj_y"]=df["doj"].dt.year  #year of joining
```

```python
df["dol_y"]=df["dol"].dt.year  #year of leaving
```

```python
df["dob_y"]=df["dob"].dt.year   #year of birth
```

```python
df["exp"] = df["dol_y"]-df["doj_y"]    #experience
```

### Univariate analysis Non-visual

```python
df_num = df.select_dtypes(include=["int", "float"])
```

```python
df_num.head()
```

```python
df_cat = df.select_dtypes(include=["object"])
df_cat.head()
```

```python
        
def discrete_univariate_analysis(df_cat):
    for col_name in df_cat:
        print("*"*10, col_name, "*"*10)
        print(df_cat[col_name].agg(['count', 'nunique', 'unique']))
        print('Value Counts: \n', df_cat[col_name].value_counts())
        print()
        
```

```python
discrete_univariate_analysis(df_cat)
```

```python
def numerical_univariate_analysis(df_num):
    for col_name in df_num:
        print("*"*10, col_name, "*"*10)
        print(df_num[col_name].agg(['min', 'max', 'mean', 'median', 'std']))
        print()
```

```python
numerical_univariate_analysis(df_num)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
X.shape
```

```python
X_train_cat = X_train.select_dtypes(include = ["object"])
X_train_cat.shape
```

```python
X_test_cat = X_test.select_dtypes(include = ["object"])
X_test_cat.shape
```

```python
X_train_num= X_train.select_dtypes(include = ["int","float"])
X_train_num.shape
```

```python
X_test_num= X_test.select_dtypes(include = ["int","float"])
X_test_num.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
X_train_cat_trans= pd.DataFrame(encoder.fit_transform(X_train_cat),columns=encoder.get_feature_names_out(),
                               index=X_train_cat.index)
X_train_cat_trans.head()



```

```python
X_test_cat_trans= pd.DataFrame(encoder.transform(X_test_cat),columns=encoder.get_feature_names_out(),
                               index=X_test_cat.index)
X_test_cat_trans.head()
```

```python
X_train_trans=pd.concat([X_train_cat_trans,X_train_num],axis=1)
X_train_trans.head()
```

```python
X_test_trans=pd.concat([X_test_cat_trans,X_test_num],axis=1)
X_test_trans.head()
```

```python
# Training a model and evaluation of learning
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)

y_test_pred = classifier.predict(X_test)

print(metrics.accuracy_score(y_test, y_test_pred))
```

```python
classifier.feature_importances_
```

```python
plt.figure(figsize=(15,7))

sns.barplot(x=classifier.feature_importances_, y=X_train.columns)

plt.show()
```
